# Amazon Sales Data Cleaning and Preparation

This dataset comprises of more than a thousand actual products from the country of India that are listed on the Amazon marketplace together with their unique identification numbers. Due to the very disorganised nature of the raw data, my goal is to clean and prepare the data for analysis. I'll export the cleaned dataset to Power BI as a csv file for data visualisation in order to gain insights and make data-driven decisions.


```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
#Display first few rows
df = pd.read_csv('amazon.csv')

df.head()
```

```python
#data attributes
df.info()

#object data type signifies strings
```

```python
#list of drop columns
columns_to_drop = ['Unnamed: 16','Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20',
       'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24',
       'Unnamed: 25']
```

```python
#drop the NaN columns
df.drop(columns=columns_to_drop, inplace=True)
```

```python
#info after dropping NaN columns 
df.info()
```

```python
#print dataset number of rows and column
print('This dataset contains', df.shape[0], 'rows and', df.shape[1], 'columns')
```

```python
#few rows after dropping NaN columns 
df.head()
```

```python
#replacing symbol and comma sign in discounted price row
#converting to float type
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')
```

```python
#replacing symbol and comma sign in actual price row
#converting to float type
df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
#replacing percentage side and conversion
df['discount_percentage'] = df['discount_percentage'].str.replace('%', "").astype('float64')
df['discount_percentage'] = df['discount_percentage']/100
df['discount_percentage']
```

```python
#checking for invalid strings
df['rating'].value_counts()
```

```python
#queried invalid string
df.query('rating == "|"')
```

```python
#replacing invalid string by its rating as identified on amazon website
#Changing Datatype
df['rating'] = df['rating'].str.replace('|', '4.0').astype('float64')
```

```python
#check
df['rating'].value_counts()
```

```python
#checking rating_count column
df['rating_count'].value_counts()
```

```python
#replacing unusual strings and changed datatype
df['rating_count'] = df['rating_count'].str.replace(',','').astype('float64')
```

```python
#Checking for any missing values
df.isna().sum()
```

```python
#display column names
df.columns
```

```python
#creating new DataFrame from the dataset
df1 = df[['product_id', 'product_name', 'category', 'discounted_price',
       'actual_price', 'discount_percentage', 'rating', 'rating_count']].copy()
```

```python
#display few new columns
df1.head()
```

```python
#splitting data in column category
splitcat = df['category'].str.split('|', expand=True)
splitcat
```

```python
#renaming category column  name
splitcat = splitcat.rename(columns={0:'category_1',1:'category_2',2:'category_3'})
splitcat
```

```python
#add category 1 and 2 to the new DataFrame
df1['category_1'] = splitcat['category_1']
df1['category_2'] = splitcat['category_2']

df1
```

```python
#drop redundant category column
df1.drop(columns='category', inplace=True)

df1
```

```python
#check category_1 for unusual strings
df1['category_1'].value_counts()
```

```python
#modify category_1
df1['category_1'] = df1['category_1'].str.replace('&', ' & ')
df1['category_1'] = df1['category_1'].str.replace('OfficeProducts', 'Office Products')
df1['category_1'] = df1['category_1'].str.replace('MusicalInstruments', 'Musical Instruments')
df1['category_1'] = df1['category_1'].str.replace('HomeImprovement', 'Home Improvement')
```

```python
#check category_2 for unusual strings
df1['category_2'].value_counts()
```

```python
#modify category_2
df1['category_2'] = df1['category_2'].str.replace('&', ' & ')
df1['category_2'] = df1['category_2'].str.replace('Home', 'Home ')
df1['category_2'] = df1['category_2'].str.replace(',', ', ')
df1['category_2'] = df1['category_2'].str.replace('WearableTechnology', 'Wearable Technology')
df1['category_2'] = df1['category_2'].str.replace('NetworkingDevices', 'Networking Devices')
df1['category_2'] = df1['category_2'].str.replace('ExternalDevices', 'External Devices')
df1['category_2'] = df1['category_2'].str.replace('OfficePaperProducts', 'Office Paper Products')
df1['category_2'] = df1['category_2'].str.replace('CraftMaterials', 'Craft Materials')
df1['category_2'] = df1['category_2'].str.replace('PowerAccessories', 'Power Accessories')
df1['category_2'] = df1['category_2'].str.replace('OfficeElectronics', 'Office Electronics')
df1['category_2'] = df1['category_2'].str.replace('CarAccessories', 'Car Accessories')
df1['category_2'] = df1['category_2'].str.replace('MedicalSupplies', 'Medical Supplies')
df1['category_2'] = df1['category_2'].str.replace('GeneralPurposeBatteries', 'General Purpose Batteries')
df1['category_2'] = df1['category_2'].str.replace('BatteryChargers', 'Battery Chargers')
```

```python
#getting rid of whitespaces in product id
df1['product_id'].str.strip()
```

```python
#creating ranking for rating 
rating_score = []

for score in df1['rating']:
    if score < 2.0: rating_score.append('Poor')
    elif score < 3.0: rating_score.append('Below Average')
    elif score < 4.0: rating_score.append('Average')
    elif score < 5.0: rating_score.append('Above Average')
    elif score == 5.0: rating_score.append('Excellent')
```

```python
#creating a new column and changing the datatype
df1['rating_score'] = rating_score
df1['rating_score'] = df1['rating_score'].astype('category')
```

```python
#checking the new column
df1['rating_score'].value_counts()
```

```python
#Reodered category excluding Poor rating rank
df1['rating_score'] = df1['rating_score'].cat.reorder_categories(['Below Average', 'Average', 'Above Average', 'Excellent'], ordered = True)
```

```python
#Creating Price difference column
df1['price_diff'] = df1['actual_price'] - df1['discounted_price']
```

```python
#Column renaming
df1 = df1.rename(columns={"product_id": "Product_ID", "product_name": "Product_Name", 
                          "discounted_price":"Discounted_Price", "actual_price":"Actual_Price", 
                          "discount_percentage":"Discount_Percentage", "rating":"Rating", "rating_count":"Rating_Count", "category_1":"Category_1",
                         "category_2":"Category_2", "price_diff":"Price_Difference"}) 
```

```python
#Outcome after first set of cleaning and preparation
df1.head()
```

```python
#Saving cleaned data
df1.to_csv('Amazon Sales', index=False)
```

Second set - Reviewers Column

```python
#subsetting reviewers identity
reviewers = df[['user_id', 'user_name']]
reviewers
```

```python
#Splitting reviewers user_id
reviewers_split_id = reviewers['user_id'].str.split(',', expand=False)
reviewers_split_id
```

```python
#explode to display an id per column
reviewers_split_exp = reviewers_split_id.explode()

#resetting index
reviewers_id_ind = reviewers_split_exp.reset_index(drop=True)
reviewers_id_ind
```

```python
#Splitting reviewers user_name
reviewers_split_name = reviewers['user_name'].str.split(',', expand=False)
reviewers_split_name
```

```python
#explode to display a reviewer name per column
reviewers_name_exp = reviewers_split_name.explode()

#resetting index
reviewers_name_ind = reviewers_name_exp.reset_index(drop=True)
reviewers_name_ind
```

```python
#creating new DataFrame
df_id = pd.DataFrame(data=reviewers_id_ind)
df_name = pd.DataFrame(data=reviewers_name_ind)
```

```python
#merging new DataFrame
df2 = pd.merge(df_id, df_name, left_index=True, right_index=True)

df2
```

```python
#rename reviewers column name
df2 = df2.rename(columns={"user_id":"User_ID", "user_name":"Username"})

df2.head()
```

```python
#Saving cleaned data
df2.to_csv('Amazon Reviewers', index=False)
```
